```python
print("""Which Operation you want to perform :-
        options are add/mul/sub/div""")
opt=str(input("Enter the operation name : "))
num1=eval(input("Enter the first number : "))
num2=eval(input("Enter the second number : "))
if(opt=='add'):
    print("Addition are : ", num1+num2)
elif(opt=='mul'):
    print("Multification are : ", num1*num2)
elif(opt=='sub'):
    if(num1>num2):
        print("Subtraction are : ", num1-num2)
    else:
        print("Subtraction are : ", num2-num1)
elif(opt=='div'):
    if(num1>num2):
        print("Division of number are : ", num2/num1)
    else:
        print("Div are : ", num1/num2)
else:
    print("Wrong operation")
```
